# Part 1: Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('XTern 2024 Artificial Intelegence Data Set - Xtern_TrainData.csv')
```

```python
# Get list of values in each column and how much they occur.
for col in df:
  print(str(df[col].value_counts()) + "\n")
```

```python
# Set helper variables for each program section
# Set colors for each food item
order_colors = {
    "Sugar Cream Pie": "C1",
    "Indiana Pork Chili": "C2",
    "Cornbread Hush Puppies": "C3",
    "Sweet Potato Fries": "C4",
    "Ultimate Grilled Cheese Sandwich (with bacon and tomato)": "C5",
    "Indiana Buffalo Chicken Tacos (3 tacos)": "C6",
    "Indiana Corn on the Cob (brushed with garlic butter)": "C7",
    "Breaded Pork Tenderloin Sandwich": "C8",
    "Fried Catfish Basket": "C9",
    "Hoosier BBQ Pulled Pork Sandwich": "C10"
}
unique_orders = df['Order'].unique()
```

```python
# Create graph showing how orders are distributed over time

# Group data based on time and order to get data for graph
time_gropued_df = df.groupby(['Time', 'Order']).size().unstack().fillna(0)

fig, ax = plt.subplots(figsize=(6, 6))

# Generate bars for each food item
bottom = np.zeros(len(time_gropued_df))
for i, order in enumerate(unique_orders):
    values = time_gropued_df[order]
    ax.bar(time_gropued_df.index, values, label=order, bottom=bottom, color=order_colors[order])
    bottom += values

# Set labels and axis
ax.set_xlabel('Order Time')
ax.set_ylabel('Number of Orders')
ax.set_title('Order Times for Food Items')
ax.legend(title='Orders', bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```

# Part 3: Model

## Process
The given problem is a multi-class classification problem, where each entity has to be classified into a specific order group. Given this problem definition, the mostly categorical features, the small number of features, and how the classes seem to split between certain groups of each category, a decision tree seems to be the best model for this problem. Decision trees also have fast training times, making the model scalable and easier to retrain with new data. I will prepare the data by encoding the categorical data with label encoding. Afterwards, I will split the data between training and test data to prevent overfitting. Using the training data I will train a decision tree model and save it using pickle. Afterwards I will get an accuracy score for the model.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pickle
```

```python
def process_data(dataframe, columns):
  label_encoder = LabelEncoder()
  for column in columns:
    dataframe[f'encoded_{column}'] = label_encoder.fit_transform(df[column])

  return dataframe
```

```python
processed_df = process_data(df, ['Year', 'University', 'Major', 'Order'])
```

```python
X = processed_df[['encoded_Year', 'encoded_University', 'encoded_Major', 'Time']]
y = processed_df['encoded_Order']
X_train, X_val, y_train, y_val = train_test_split(X, y)
```

```python
def create_model(X_train, y_train):
  model = DecisionTreeClassifier()
  model.fit(X_train, y_train)
  with open('decision_tree_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
create_model(X_train, y_train)
```

```python
model = None
with open('decision_tree_model.pkl', 'rb') as file:
    model = pickle.load(file)
predicted = model.predict(X_val)
print(accuracy_score(y_val, predicted))
```
